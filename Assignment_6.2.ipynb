```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmx
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
```

```python
#Pie Chart
pie_plot = titanic.groupby('sex').size().plot(kind='pie', legend=True)
plt.show()
```

```python
#Scatter Plot
x=titanic['age']
y=titanic['fare']
```

```python
# Get unique names of sex
uniq = list(set(titanic['sex']))

# Set the color map to match the number of sexes
z = range(1,len(uniq))
hot = plt.get_cmap('hot')
cNorm  = colors.Normalize(vmin=0, vmax=len(uniq))
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=hot)
```

```python
for i in range(len(uniq)):
    indx = titanic['sex'] == uniq[i]
    plt.scatter(x[indx], y[indx], s=15, color=scalarMap.to_rgba(i), label=uniq[i])
```

```python
plt.xlabel('Age')
plt.ylabel('Fair Paid')
plt.title('Age VS Fare')
plt.legend(loc='upper left')
plt.show()
```
